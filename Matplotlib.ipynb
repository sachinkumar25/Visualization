```python
import math

classes = ['9th','10th','11th','12th','College']
students = [123,312,412,523,690]

plt.bar(classes, students)
plt.title('The amount of students in each class')
plt.xlabel('Class')
plt.ylabel('# of Students')
```

```python
import math

classes = ['9th','10th','11th','12th','College']
students = [123,312,412,523,690]
explode = [0,0,0,0.1,0]

plt.pie(x = students, labels = classes, explode = explode)
plt.axis('equal')
plt.legend(loc= 'upper right')
plt.show()
```
